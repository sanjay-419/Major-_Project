```python
import pandas as pd
import numpy as np
from sklearn.multioutput import MultiOutputClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
n=99000
data.drop(data.tail(n).index,
        inplace = True)
```

```python
data1=pd.read_csv('Heart.csv')
```

```python
data1
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
sex = le.fit_transform(data1['sex'])
data1['sex']=sex
cp= le.fit_transform(data1['cp'])
fbs=le.fit_transform(data1['fbs'])
rest_ecg=le.fit_transform(data1['rest_ecg'])
exng=le.fit_transform(data1['exng'])
slope=le.fit_transform(data1['slope'])
smoking_history=le.fit_transform(data['smoking_history'])
data1['sex']=sex
data1['cp']=cp
data1['fbs']=fbs
data1['rest_ecg']=rest_ecg
data1['exng']=exng
data1['slope']=slope
data['smoking_history']=smoking_history
```

```python
data1.rename(columns={'target': 'heart_disease'}, inplace=True)
```

```python
data1
```

```python
datahd=data1.merge(data, on='heart_disease',how='right')
```

```python
n=100000
datahd.drop(datahd.tail(n).index,
        inplace = True)
 
```

```python
x = datahd.drop(['heart_disease', 'diabetes'], axis=1)
```

```python
import pickle
from sklearn.model_selection import cross_val_score, train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score , precision_score , recall_score
```

```python
from sklearn.model_selection import train_test_split ,RandomizedSearchCV
```

```python
X_train,x_test,Y_train,y_test=train_test_split(x,y,random_state=20,test_size=0.40)
```

ALOGRITHM BUILDER

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1= RandomForestClassifier(n_estimators=100, random_state=42

)
model = MultiOutputClassifier(model1)
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
acc = accuracy_score(y_test, model.predict(x_test)) * 100
```

```python
acc
```

```python
precision = precision_score(y_test, y_pred, average='macro')
```

```python
recall = recall_score(y_test, y_pred, average='macro')
```

```python
file_name='HD.pkl'
pickle.dump(model,open(file_name,'wb'))
```

```python
loaded=pickle.load(open(file_name,'rb'))
result=loaded.score(x_test,y_test)
```
