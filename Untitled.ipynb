```python
import pandas as pd 
import pickle
from sklearn.model_selection import cross_val_score, train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Heart.csv")
data.head(10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data.info()
```

```python
data.describe().T
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
sex = le.fit_transform(data['sex'])
cp= le.fit_transform(data['cp'])
fbs=le.fit_transform(data['fbs'])
rest_ecg=le.fit_transform(data['rest_ecg'])
exng=le.fit_transform(data['exng'])
slope=le.fit_transform(data['slope'])
```

```python
data.drop("slope", axis=1, inplace=True)
```

```python
data['sex']=sex
data['cp']=cp
data['fbs']=fbs
data['rest_ecg']=rest_ecg
data['exng']=exng
data['slope']=slope
```

```python
data.head()
```

```python
y=data["target"]
```

```python
y
```

```python
x=data.drop('target',axis=1)
```

```python
x.tail()
```

```python
X_train,X_test,Y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=20)
```

```python
LR=LogisticRegression()
```

```python
model=LR.fit(X_train,Y_train)
```

```python
y_pred=LR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , precision_recall_curve , auc
```

```python
acc = accuracy_score(y_test, LR.predict(X_test)) * 100
```

```python
acc
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
auc_score = auc(recall, precision)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'Precision-Recall Curve (AUC = {auc_score:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

```python
file_name='model.pkl'
pickle.dump(model,open(file_name,'wb'))
```

```python
loaded=pickle.load(open(file_name,'rb'))
result=loaded.score(X_test,y_test)
```
